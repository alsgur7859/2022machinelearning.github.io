## define the feature functions

- feature vector is defined by $(1, f_1(x, y), f_2(x, y), \cdots, f_{k-1}(x, y)) \in \mathbb{R}^k$

```python
def compute_feature1(point):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    #
    feature = np.ones((5,point.shape[0]))
    feature[1] = np.transpose(point)[0]
    feature[2] = np.transpose(point)[1]**2
    feature[3] = np.transpose(point)[0]**2
    feature[4] = np.transpose(point)[0] + np.transpose(point)[1]
    feature = np.transpose(feature)
    
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return feature
```

```python
def compute_feature2(point):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks

    feature = np.ones((7,point.shape[0]))
    feature[1] = np.transpose(point)[0]
    feature[2] = np.transpose(point)[0]**2
    feature[3] = np.transpose(point)[0]**3
    feature[4] = np.transpose(point)[0]**4
    feature[5] = np.transpose(point)[1]
    feature[6] = np.transpose(point)[1]**2
    feature = np.transpose(feature)
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return feature
```

## define the linear regression function

- $\theta = (\theta_0, \theta_1, \cdots, \theta_{k-1}) \in \mathbb{R}^k$
- feature = $(1, f_1(x, y), \cdots, f_{k-1}(x, y)) \in \mathbb{R}^k$

```python
def compute_linear_regression(theta, feature):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    value = np.inner(np.transpose(theta), feature)

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return value
```

## define sigmoid function with input

- $z \in \mathbb{R}$

```python
def sigmoid(z):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #

    value = 1 / (1 + np.exp(-z))
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return value 
```

## define the logistic regression function

- $\theta = (\theta_0, \theta_1, \cdots, \theta_{k-1}) \in \mathbb{R}^k$
- feature $= (1, f_1(x, y), \cdots, f_{k-1}(x, y) \in \mathbb{R}^k$

```python
def compute_logistic_regression(theta, feature):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #

    value = sigmoid(compute_linear_regression(theta, feature))
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return value
```

## define the residual function

- $\theta = (\theta_0, \theta_1, \cdots, \theta_{k-1}) \in \mathbb{R}^k$
- feature $= (1, f_1(x, y), \cdots, f_{k-1}(x, y) \in \mathbb{R}^k$
- label $= l \in \{0, 1\}^k$

```python
def compute_residual(theta, feature, label):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    
    residual = (label) * (-np.log(compute_logistic_regression(theta, feature) + 0.0000001)) \
    + (1 - label) * (-np.log(1 - compute_logistic_regression(theta, feature) + 0.0000001))

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return residual
```

## define the loss function for the logistic regression

- $\theta = (\theta_0, \theta_1, \cdots, \theta_{k-1}) \in \mathbb{R}^k$
- feature $= (1, f_1(x, y), \cdots, f_{k-1}(x, y) \in \mathbb{R}^k$
- label $= l \in \{0, 1\}^k$

```python
def compute_loss(theta, feature, label):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    loss = compute_residual(theta, feature, label).sum() / compute_residual(theta, feature, label).size

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return loss
```

## define the gradient of the loss with respect to the model parameter $\theta$

- $\theta = (\theta_0, \theta_1, \cdots, \theta_{k-1}) \in \mathbb{R}^k$
- feature $= (1, f_1(x, y), \cdots, f_{k-1}(x, y) \in \mathbb{R}^k$
- label $= l \in \{0, 1\}^k$

```python
def compute_gradient(theta, feature, label):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    gradient = np.matmul((compute_logistic_regression(theta, feature) - label.reshape(-1,)), feature) / feature.shape[0]

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return gradient
```

## compute the accuracy of the prediction for point with a given model parameter

```python
def compute_accuracy(theta, feature, label):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    label_predict = np.logical_not(compute_logistic_regression(theta, feature) < 0.5)
    label_wrong = np.logical_xor(label, label_predict)
    # print(np.logical_xor(label, label_predict))
    accuracy = (label.size - np.count_nonzero(label_wrong)) / label.size
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return accuracy
```

## initialize the gradient descent algorithm

```python
data1_number_iteration   = 10000
data2_number_iteration   = 30000

data1_learning_rate   = 0.7
data2_learning_rate   = 1.0

data1_number_feature    = 5
data2_number_feature    = 7 

theta1  = np.zeros(data1_number_feature)
theta2  = np.zeros(data2_number_feature)

data1_loss_iteration  = np.zeros(data1_number_iteration)
data2_loss_iteration  = np.zeros(data2_number_iteration)

data1_accuracy_iteration    = np.zeros(data1_number_iteration)
data2_accuracy_iteration    = np.zeros(data2_number_iteration)
```

## run the gradient descent algorithm to optimize the loss function with respect to the model parameter

```python
for i in range(data1_number_iteration):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    # print(data1_point)
    # print(compute_feature1(data1_point))
    # print(compute_linear_regression(theta1, compute_feature1(data1_point)))
    # print(compute_logistic_regression(theta1, compute_feature1(data1_point)))
    # print(data1_label)
    # print(compute_gradient(theta1, compute_feature1(data1_point), data1_label))
    # print(theta1 - data1_learning_rate * compute_gradient(theta1, compute_feature1(data1_point), data1_label))
    # print(compute_loss(theta1, compute_feature1(data1_point), data1_label))
    # print(compute_accuracy(theta1, compute_feature1(data1_point), data1_label))
    theta1      = theta1 - data1_learning_rate * compute_gradient(theta1, compute_feature1(data1_point), data1_label)
    loss1       = compute_loss(theta1, compute_feature1(data1_point), data1_label)
    accuracy1   = compute_accuracy(theta1, compute_feature1(data1_point), data1_label)

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    data1_loss_iteration[i]     = loss1
    data1_accuracy_iteration[i] = accuracy1

data1_theta_optimal = theta1
```

```python
for i in range(data2_number_iteration):
    
    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    theta2      = theta2 - data2_learning_rate * compute_gradient(theta2, compute_feature2(data2_point), data2_label)
    loss2       = compute_loss(theta2, compute_feature2(data2_point), data2_label)
    accuracy2   = compute_accuracy(theta2, compute_feature2(data2_point), data2_label)

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    data2_loss_iteration[i]     = loss2
    data2_accuracy_iteration[i] = accuracy2

data2_theta_optimal = theta2
```
